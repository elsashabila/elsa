```python
#Simpan hasil operasi ke dalam variabel
harga_total_kojek = 3000*3

print(harga_total_kojek)
```

```python
#Simpan nilai operand dan hasil operasi ke dalam variabel
harga_cilok = 10000
jumlah_porsi_cilok = 5
harga_total_cilok =10000*5

print(harga_total_cilok)
```

```python
uang_yang_diterima = 100000
uang_kembalian = uang_yang_diterima - harga_total_cilok

print(uang_kembalian)
```
