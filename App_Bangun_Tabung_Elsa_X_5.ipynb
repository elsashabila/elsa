```python
#Program menghitung volume Tabung

#Meminta input dari pengguna
pi = (22/7)
jari = float(input("Masukkan jari-jari tabung (cm):"))
tinggi = float(input("Masukkan tinggi tabung (cm):"))

#Menghitung volume tabung
volume_tabung = pi*jari**2*tinggi

#Menampilkan hasil
print(f"volume tabung adalah: {volume_tabung} cm^3.")
```
